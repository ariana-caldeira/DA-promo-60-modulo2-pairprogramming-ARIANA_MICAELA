3. Usando "information_schema" en la conexión a base de datos, extraer los nombres de las columnas de la tabla "customers" de la bases de datos "northwind".

```python
cnx = mysql.connector.connect(user='root', password='Fefs1512',
                              host='127.0.0.1',
                              database='Northwind')

mycursor = cnx.cursor()

mycursor.execute("""SELECT COLUMN_NAME
        FROM information_schema.COLUMNS
        WHERE TABLE_SCHEMA = 'Northwind'
          AND TABLE_NAME = 'customers';""")

myresult = mycursor.fetchall()
cnx.close()

print(myresult)
```
